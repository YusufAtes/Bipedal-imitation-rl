```python
import pandas as pd
import matplotlib.pyplot as plt
logger = pd.read_csv('ppo_upt_64_64_softupdate.csv')
rewards = logger['Total Reward']
steps = logger['Termination Step']
max_reward_ind = rewards.idxmax()
max_reward = rewards[max_reward_ind]
max_reward_step = steps[max_reward_ind]
print(f'Max reward: {max_reward} at step {max_reward_step}')
plt.plot(rewards)
```
