```python
import os
import numpy as np
import time
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("ppo_256_256/PPO_13/demo_avg_results_0.1_0.2.csv")

print(df.head())

# get the total episode number from rows
total_episodes = df.shape[0]

# Get the success rate from the number of successful episodes  
success_rate = df['Success'].mean() * 100
print(f"Total Episodes: {total_episodes}")
print(f"Success Rate: {success_rate:.2f}%")

# find the unique Speed Values
unique_speeds = df['Speed Value'].unique()

# Plotting the success rate against speed values
plt.figure(figsize=(10, 6))
plt.plot(unique_speeds, df.groupby('Speed Value')['Success'].mean() * 100, marker='o')
plt.title('Success Rate vs Speed Value')
plt.xlabel('Speed Value')
plt.ylabel('Success Rate (%)')
plt.xticks(unique_speeds)
plt.grid()

# Plot the Mean Speed against Speed Values only from the successful episodes
plt.figure(figsize=(10, 6))
plt.plot(unique_speeds, df[df['Success'] == 1].groupby('Speed Value')['Mean Speed'].mean(), marker='o', color='orange')
plt.title('Mean Speed vs Speed Value')
plt.xlabel('Speed Value')
plt.ylabel('Mean Speed')
plt.xticks(unique_speeds)
plt.grid()
# Show the plots
```
